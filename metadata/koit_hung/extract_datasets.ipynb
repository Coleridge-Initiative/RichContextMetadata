```python
import pandas as pd
import json
```

```python
loc = 'JSON formula.xlsx'
df = pd.read_excel(loc, sheet_name = 'Data',header = 0)
```

```python
kh_dataset_list= df[['alt_title','title']].to_dict('records')
```

```python
[k.update({'alt_title':[k['alt_title']]}) for k in kh_dataset_list]
```

```python
with open('kh_datasets.json', 'w') as outfile:
    json.dump(kh_dataset_list, outfile,indent = 2)
```
