```python
import json
import numpy as np
import pandas as pd
```

```python
filename = '/Users/andrewnorris/RCHuman/manual/partitions/corpus.0_1_5.json'
```

```python
with open(filename, 'r') as f_in:
    data = json.load(f_in)
```

```python
entry_list= list()
for i in data:
    entry = dict()
    entry['title'] = i['title']
    original=dict()
    original['pdf'] = i['manual']['pdf']
    original['doi'] = i['manual']['doi']
    original['url'] = i['manual']['url']
    original['journal'] = i['manual']['journal']
    dataset = list(i['manual']['datasets'])
    entry['original']=original
    entry['dataset'] = dataset
    entry_list.append(entry)
print(entry_list)
```

```python
with open('bbk_corpus0_1_5_partition.json', 'w') as f_out:
    json.dump(entry_list, f_out, indent=4, sort_keys=True, ensure_ascii=False)
```
