```python
entry = dict()
entry_list= list()
for i in data:
    entry['title'] = i['title']
    original=dict()
    original['pdf'] = i['manual']['pdf']
    original['doi'] = i['manual']['doi']
    original['url'] = i['manual']['url']
#    original['journal'] = i['manual']['journal']
    dataset = list(i['manual']['datasets'])
    entry['original']=original
    entry['dataset'] = dataset
    entry_list.append(entry)
print(entry_list)

```

```python
with open('newfile.json', 'w') as f_out:
    json.dump(entry_list, f_out, indent=4, sort_keys=True, ensure_ascii=False)
```
