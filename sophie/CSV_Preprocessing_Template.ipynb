### Preprocessing metadata

This is a template notebook for preprocessing metadata 

Preprocess goes as follows:
1. Walk through all the CSV files. Open them and process.
2. Check if `journal` column values are in  wrong format, pass if it's blank.
3. Output the csv with new encoding in the original directory to a new file (by default) or overwrite the original one.

```python
import pandas as pd
import os
pd.set_option('display.max_colwidth', -1)
```

#### Change the file you want to preprocess here!

```python

parent_folder = '../RichContextMetadata/sophie'
```

```python
# Unify all the journal names
def Unified_Journal(df):
    i = 0 
    while i < len(df):
        if type(df['journal'].iloc[i]) == float:
            i = i + 1
            continue
        if '{' in df['journal'].iloc[i]:
            df['journal'].iloc[i] = df['journal'].iloc[i][33:-2]
        i = i + 1
    return df
# Maybe further I should update my algorithm instead of using brutal force, but it seems to work for now.
```

```python
# Out put csv with new encoding. 
def Out_Put(df, file_path, overwrite=False):
    if overwrite == True:
        df.to_csv(file_path, encoding = 'utf-8-sig')
    else:
        new_path = file_path[:-3]+'_jason.csv'
        df.to_csv(new_path, encoding = 'utf-8-sig')
```

```python
# Walk through and process all the csv in the file. 
# We could add more functions in the future.
def Preprocess(path):
    content = os.walk(parent_folder)  
    for path,dir_list,file_list in content:  
        for file_name in file_list:
            f = os.path.join(path, file_name)
            if f[-4:] == '.csv':
                df = pd.read_csv(f, encoding = 'UTF-8')
                df = Unified_Journal(df)
                Out_Put(df, f)
    print('Done!')
```
